```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay

from ood_plotting import plot_ood_scores

SAVE_PLOTS = True
#plt.rcParams['figure.figsize'] = [4,2.5]
plt.rcParams['figure.figsize'] = [4,2.5]
```

# Basic usage

```python
# Generate scores for ID and OOD samples
id_data = np.random.normal(0,0.2,500)
ood_data = np.random.normal(0.4,0.1,500)

# Compute standard OOD metrics and plot histogram + threshold curve for AUTC
plot_ood_scores(id_data,ood_data)
```

# Comparison of 9 imaginary models (Figs. 3,4,6,7)

```python
for row in [1,2,3]:

    res = []
    for model in [1,2,3]:
        print(f"-------------- row {row}, model {model} --------------")
        id_data = np.load(f"synthetic_data/row{row}-{model}-id.npy",allow_pickle=True)
        ood_data = np.load(f"synthetic_data/row{row}-{model}-ood.npy",allow_pickle=True)

        res.append(plot_ood_scores(id_data,ood_data,save=SAVE_PLOTS,model_name=f"model {(row-1)*3+model}",return_metrics=True, plot_threshs=False))

    fig, axs = plt.subplots(1, 1, figsize=(2,2))
    for c in res:
        display = RocCurveDisplay.from_predictions(c["labels"], c["preds"],
                                      name=c["name"],ax=axs,linewidth=1,label=c["name"])

    axs.set_ylabel("True Positive Rate")
    axs.set_xlabel("False Positive Rate")
    axs.set_xticks([0, 0.5, 1])
    axs.set_yticks([0, 0.5, 1])
    plt.tight_layout(pad=0)

    if SAVE_PLOTS: plt.savefig(f"ood_plots/roc-curve-row{row}.pdf")

    plt.show()
```
